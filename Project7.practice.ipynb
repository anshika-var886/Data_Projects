Q.1 What is the relationship between Rate of Cognizable Crimes and Chargesheet Rate in the year 2022?

```python
data.dtypes
```

```python
max_crimes_state = data.sort_values('Rate of Cognizable Crimes (IPC) (2022)', ascending= False).head(5)
max_crimes_state
```

```python
chargsheet_states = data.sort_values('Chargesheeting Rate (2022)', ascending= False).head(5)
chargsheet_states 
```

```python
concated_data= pd.concat([max_crimes_state, chargsheet_states], ignore_index= True, axis=0)
concated_data = concated_data.drop_duplicates()
concated_data
```

```python
fig, axes= plt.subplots(figsize=(15,4))
ax2= axes.twinx()
sns.barplot(x='State/UT', y= 'Rate of Cognizable Crimes (IPC) (2022)', data= concated_data, ax=axes,
             hue='State/UT', legend= False, palette= 'viridis')
sns.lineplot(x='State/UT', y= 'Chargesheeting Rate (2022)', data= concated_data, ax=ax2,
             color= 'blue', linestyle='--', marker= 'o')
plt.title('High cognizable crimes or chargesheeting rate wise states and their comparison')
plt.show()
```

We find Delhi being the stand out in terms of high crime rate and low chargesheeting rate. MP, Puducherry, WB, and Guj have sort of similar Chargesheeting rate. Kerala stands out by handling the case with regards to having more than 90% chargesheeting rate and other states can reflect upon kerala's policies in crime handling. 

```python
melted= pd.melt(data, id_vars= ['State/UT'], value_vars=['2020', '2021', '2022'],
                var_name= 'Reporting_Year', value_name= 'Crimes_Reported')
melted
```

Q.2 How the change varies in reporting of crime in each year and in each state?

```python
melted= melted.sort_values(['State/UT', 'Reporting_Year'])
melted['pct_change']= melted.groupby('State/UT')['Crimes_Reported'].pct_change()
melted
```

```python
melted['pct_change'] = melted['pct_change'].fillna(0)
```

```python
sns.barplot(x='pct_change', y='State/UT', hue='Reporting_Year', data=melted)
plt.title('Reporting crime changes in each state in successive years')
plt.show()
```

Delhi has seen the highest % change from the year 2021-22 with the rise of over 1.5% no. of case reported. The COVID effect is variably distributed. However we do see decline in pct-change is low (maximum till -0.5) as compared to the increase(1.5%). Kerala  predominantly see the largest change or increase in the case after 2021. 

Q.3 What is the crime rate in the year 2022 prevailing in each region. Compared it to the Indian average.

```python
df= data.copy()
df['Crime Rate 2022'] = df['2022']/df['Mid-Year Projected Population (in Lakhs) (2022)']
df = df.sort_values('Crime Rate 2022', ascending= False)
```

```python
mean_rate= df['Crime Rate 2022'].mean()
plt.figure(figsize=(15,7))
graph= sns.barplot(x= 'Crime Rate 2022', y='State/UT', hue= 'State/UT', legend= False, palette= 'mako', data= df)
graph.axvline(mean_rate, color='red', linestyle='--', label='Average Crime Rate in India 2022')
plt.text(x=mean_rate + 20,
         y=10, 
         s=f'Average: {mean_rate:.2f}', 
         color='red',
         ha='left',
         va='center')
plt.title('States/UTs Crime Rate in 2022')
plt.show()
```

There are approximately 10 states that are driving the average. The crime rate is independent of the States/UTs population. Even small states like Mizoram has a crime rate above the indian average. Problem is serious in Delhi and Haryana with high crime rate but low chargesheeting rate as visualized above.

Q.4 Analyze the distribution of crimes reported in each year as well as of the mid year projected population.

```python
from matplotlib.ticker import FuncFormatter
formatter = FuncFormatter(lambda x, p:format(int(x), ','))
col1 = ['2020', '2021', '2022']
col2= ['Mid-Year Projected Population (in Lakhs) (2022)']
min_x= min(data[col].min() for col in col1)
max_x= max(data[col].max() for col in col1)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 5), sharex=False, sharey=False)
axes = axes.flatten()
for i, col in enumerate(col1):
    sns.histplot(x=col, data=data, kde=True,ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].xaxis.set_major_formatter(formatter)
    axes[i].set_xlim(min_x, max_x)
    axes[i].tick_params(axis='x', rotation=90)

for col in col2:
    sns.histplot(x=col, data=data, kde=True, ax= axes[3])
    axes[3].set_title(f'Distribution of {col}', fontsize=12)
plt.tight_layout()
plt.show()
```

The data is skewed and not properly distributed. There are presence of outliers in the distribution of 2020 and in Mid year projected population. Most of the crimes reported are below 1 lakh in each year. 

Q.5 How is per capita crime rate distributed in high population states and low population states? 

```python
df = data.copy()
df['Per_capita_crime_rate'] = (df['Rate of Cognizable Crimes (IPC) (2022)']/100000)*100
```

```python
min_pop = data['Mid-Year Projected Population (in Lakhs) (2022)'].min()
low_pop = data[data['Mid-Year Projected Population (in Lakhs) (2022)'] == min_pop]['State/UT'].values[0]
low_pop
```

```python
max_pop = data['Mid-Year Projected Population (in Lakhs) (2022)'].max()
high_pop = data[data['Mid-Year Projected Population (in Lakhs) (2022)'] == max_pop]['State/UT'].values[0]
high_pop
```

```python
populous_states = df.sort_values('Mid-Year Projected Population (in Lakhs) (2022)', ascending= False).head(5)
least_populous_states = df.sort_values('Mid-Year Projected Population (in Lakhs) (2022)').head(5)
```

```python
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize

cmap = 'YlGnBu' #setting the color
norm = Normalize(vmin=df['Mid-Year Projected Population (in Lakhs) (2022)'].min(), #creating the range for the color scale
                 vmax=df['Mid-Year Projected Population (in Lakhs) (2022)'].max())
sm = ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([]) 

fig, ax= plt.subplots(nrows=2, ncols= 1, figsize=(15,5), sharex=True)

highly_populous = [sm.to_rgba(p) for p in populous_states['Mid-Year Projected Population (in Lakhs) (2022)']] #giving color range to populous states 
sns.barplot(x='Per_capita_crime_rate', y= 'State/UT', hue='State/UT', legend=False, data= populous_states,  
            palette= highly_populous, ax=ax[0])

less_populous = [sm.to_rgba(p) for p in least_populous_states['Mid-Year Projected Population (in Lakhs) (2022)']] #same to the least populous ones
sns.barplot(x= 'Per_capita_crime_rate', y= 'State/UT', hue='State/UT', legend=False, data= least_populous_states,
            palette= less_populous, ax=ax[1])
cbar = fig.colorbar(sm, ax=ax, orientation= 'vertical', pad=0.1) #calling the colorbar
cbar.set_label('Mid year projected population of the states')
plt.title('Per capita crime rate in high and low populous states')
plt.show()
```

We find that top least populous states have generally the low per capita crime rate, though Ladakh is the one leading the trend. In high populous states MP has highest per capita crime rate. However we should be cautious in not generalizing it as we do see states like Mizoram having high crime rate, despite in low population.
